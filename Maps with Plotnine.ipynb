```python
us_states = GeoDataFrame.from_file('http://www2.census.gov/geo/tiger/GENZ2017/shp/cb_2017_us_state_5m.zip')
```

```python
us_states
```

```python
us_states_50 = us_states.query("STUSPS not in ['DC', 'VI', 'PR', 'AS', 'GU', 'MP']")
len(us_states_50)
```

```python
us_states_50
```

```python
us_states_continental = us_states_50.query("STUSPS not in ['HI']")
len(us_states_continental)
```

```python
us_states_contiguous = us_states_continental.query("STUSPS not in ['AK']")
len(us_states_contiguous)
```
